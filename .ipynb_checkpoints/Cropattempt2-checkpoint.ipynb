```python
from osgeo import gdal
from scipy.misc import bytescale
outvrt = '/vsimem/stacked.vrt' #/vsimem is special in-memory virtual "directory"
outtif = 'stacked.tif'
tifs = ['SEJB04.tif', 'SEJB03.tif', 'SEJB02.tif'] 

outds = gdal.BuildVRT(outvrt, tifs, separate=True)
outds = gdal.Translate(outtif, outds)
```

```python
def tiff_to_array(tiff_file, lon_offset_px=0, lat_offset_px=0):
    gdo = gdal.Open(tiff_file)
    tiff_array = gdo.ReadAsArray(lon_offset_px,lat_offset_px)
    tiff_array = np.array(bytescale(tiff_array, low=0, high=255), dtype=np.int16)
    gdo = None
    return tiff_array
red = tiff_to_array('EH_B04.tif')
green = tiff_to_array('EH_B03.tif')
blue = tiff_to_array('EH_B02.tif')
import numpy as np
from scipy.misc import bytescale
from skimage import exposure
from matplotlib import pyplot as plt

def create_composite(red_band, green_band, blue_band):
    img_dim = red.shape
    img = np.zeros((img_dim[0], img_dim[1], 3), dtype=np.float)
    img[:,:,0] = red
    img[:,:,1] = green
    img[:,:,2] = blue
    p2, p98 = np.percentile(img, (2, 98))
    img_rescale = exposure.rescale_intensity(img, in_range=(p2, p98))
    return bytescale(img_rescale)

image = create_composite(red, green, blue)
plt.imshow(image)
```
